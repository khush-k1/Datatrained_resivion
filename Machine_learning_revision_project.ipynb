```python
# Q3. If any country is invalid change it to valid country and also fill nan

df['Country'].unique()
```

we can see UAS looks like its a typo error, so we change it to USA. also we have nan so we can fill it with mode of this column as it is an categorical column.

```python
# replacing UAS to USA
df['Country'].replace({'UAS': 'USA'}, inplace= True)
df
```

```python
# replacing Nan with mode value
df['Country'].fillna(value= df['Country'].mode()[0], inplace=True)

df
```
