```python
# Q3. If any country is invalid change it to valid country and also fill nan

df['Country'].unique()
```

we can see UAS looks like its a typo error, so we change it to USA. also we have nan so we can fill it with mode of this column as it is an categorical column.

```python
# replacing UAS to USA
df['Country'].replace({'UAS': 'USA'}, inplace= True)
df
```

```python
# replacing Nan with mode value
df['Country'].fillna(value= df['Country'].mode()[0], inplace=True)

df
```

# Q4. Add total income column which is submission of Income.1 and Income.2


let's start with Income.1 as we have clear Nan and blank $ in it.

```python
# checking values of Income.1
df['Income.1'].unique()
```

we can see their is $ sign which we need to remove in order to make it an int column.

```python
df['Income.1'].dtype
```

```python
#replacing $ 
df['Income_1']= df['Income.1'].str.replace('$','')

```

```python
# Since we have two just $ sign which we replace with black space.. We just assume they don't have ny Income_1 so assign 0.
df['Income_1']= df['Income_1'].replace('','0')

df['Income_1']
```

```python
#converting INcome_1 to numerical form
df['Income_1']= df['Income_1'].astype(float)

df['Income_1'].dtypes
```

```python
# replacing Nan value with mean 
df['Income_1'].fillna(value= df['Income_1'].mean(), inplace=True)

#chceking dataset
df
```

```python
# let's drop Income.1 column as werecover this column in Income_1.

df.drop('Income.1', axis= 1, inplace= True)

df
```

```python
# let's fill Nan value for Income.2 column with its mean.

df['Income.2'].fillna(value= df['Income.2'].mean(), inplace=True)

df
```

```python
# Adding Total Income column

df['Total_Income']= df['Income_1'] + df['Income.2']

df
```
